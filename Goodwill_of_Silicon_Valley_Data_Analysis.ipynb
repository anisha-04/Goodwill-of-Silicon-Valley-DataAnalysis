**Goodwill Silicon Valley is a California nonprofit public benefit organization** dedicated to helping people overcome employment barriers and transform lives through vocational training and environmental stewardship. With 18 retail stores, an online store, and recycling operations, they provide a wide range of services to the community. Their mission is to raise living standards and improve lives through social enterprise. **Founded in 1928**, they are part of the global Goodwill Industries International federation, making a significant impact on employment opportunities and economic independence.



The data is collected from the organization's Audited Financial statements from year 2022. The link of all the audited financial statements from all years can be found here: https://goodwillsv.org/about-us/financial-statement/.

We are doing data analysis on the 'Audited Consolidated Financial statement and Supplementary Information' of Goodwill of Silicon Valley organization from 'December 31, 2022 and 2021. This data can be found here: https://goodwillsv.org/wp-content/uploads/2023/05/Goodwill-of-Silicon-Valley-2022.pdf




Key Insights obtained from Financial Audit report:

*   Total revenue for 2022 was USD 73.4 million, up slightly from USD 73.0 million in 2021. The biggest revenue sources were retail programs at USD 40.2 million and donated goods at USD 14.2 million.
*   Total expenses in 2022 were USD 66.8 million, up from USD 60.9 million in 2021. The largest expenses were the cost of donated goods sold at USD 14.3 million and retail program expenses at USD 37.2 million.
*   Net assets increased by USD 7.0 million in 2022 compared to a USD 12.3 million increase in 2021. The change was primarily driven by operating activities.
*   The organization received USD 4.3 million in federal grant funding in 2022, compared to USD 5.3 million in 2021. Major federal programs included the American Rescue Plan Act, Housing Trust Fund, and Supplemental Nutrition Assistance Program.
*   Cash and investments totaled USD 45 million at the end of 2022, providing good liquidity. This was up from USD 36.2 million at the end of 2021.







Let's see the difference between the Major revenue and expense categories from years 2022 and 2021.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Major revenue and expense categories:

revenue_2021 = [36.4, 14.5, 8.2, 5.0, 5.4, 1.4, 1.4, 0.3]
revenue_2022 = [40.2, 14.2, 7.6, 4.7, 5.5, 1.4, 0.3, 0.1]

expenses_2021 = [33.3, 14.3, 8.7, 1.0, 3.5, 0.1]
expenses_2022 = [37.2, 14.3, 9.5, 1.1, 4.5, 0.1]

labels = ['Retail', 'Donated Goods', 'Grants', 'Operations', 'E-Commerce',
          'Contracts', 'Investments', 'Other']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, revenue_2021, width, label='2021')
rects2 = ax.bar(x + width/2, revenue_2022, width, label='2022')

ax.set_ylabel('Revenue ($ millions)')
ax.set_title('Goodwill Revenue by Category')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

2. Donated goods revenue experienced a slight decline from USD 14.5 million to USD 14.2 million between 2021 and 2022.

3. Grant revenue also decreased from USD 8.2 million to USD 7.6 million over the two-year period.

The stacked area chart was chosen as an effective means of visualizing how the composition of revenue changed over time, allowing clear insights into each category's contribution to the total revenue in each period. However, the chart does have some potential limitations. It may make it harder to observe trends in smaller categories like operations and grants. # **Conclusion**

**Conclusion of Exploratory Data Analysis of Goodwill's 2021-2022 Financial Analysis:**

Goodwill's revenue remained stable at around USD 73 million from 2021 to 2022. Retail revenue increased by approximately 10%, while grants and investment income declined. Expenses grew by about 9%, primarily driven by higher retail program costs. This led to a decline in the profit margin from 16.7% to 9.6%.

Despite the margin compression, Goodwill's assets increased from USD 59 million to USD 75 million, mainly due to significant growth in the investment portfolio. However, liabilities also rose from USD 10.2 million to USD 19.1 million, primarily because of new operating lease obligations. Net assets increased by USD 7 million in 2022, showing lower growth compared to the previous year due to lower profit margins.

On the positive side, liquid assets improved, providing around 9 months of operating cash, enhancing Goodwill's financial flexibility. Overall, the organization maintained stability and a healthy financial position but needs to monitor costs more closely in the future. Focusing on core retail programs and managing expenses will be important to sustain financial performance.
